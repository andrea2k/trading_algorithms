# Import all libraries

```python
import datetime
import time
import yfinance as yf
from dash import Dash, html, dcc
from dash.dependencies import Input, Output
import matplotlib.pyplot as plt
```

# Pull the data

```python
company = "AAPL"
now = time.strftime("%Y-%m-%d")
data = yf.download(company,start="2021-01-01", end=now)
print(data)
```

```python
data_high = data.High
data_low = data.Low
data_price = data.Open
x = data.index

plt.figure(figsize=(10,6))
plt.plot(x, data_high, label = "high")
plt.plot(x, data_low, label = "low")
plt.plot(x, data_price, label = "price")

plt.show()
```
